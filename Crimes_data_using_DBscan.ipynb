```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
```

```python
crime=pd.read_csv('/crime_data.csv')
```

```python
crime.info()
```

```python
crime2=crime.drop('Unnamed: 0',axis=1)
```

```python
crime3=pd.DataFrame(normalize(crime2),columns=crime2.columns)
```

```python
crime3
```

```python
array=crime3.values
```

```python
array
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=0.1,min_samples=3)
dbscan.fit(array)
```

```python
dbscan.labels_
```

```python
final=pd.DataFrame(dbscan.labels_,columns=['clusters'])
```

```python
crime['clusters']=final
```

```python
crime
```

```python
crime_filtered=crime.drop(crime.index[[10]],axis=0).reset_index()
```

```python
crime_filtered
```

```python
# After removing the outliers applying KMeans
```

```python
from sklearn.cluster import KMeans
```

```python
crimes=crime_filtered.drop(['Unnamed: 0','clusters'],axis=1)
```

```python
crimes1=pd.DataFrame(normalize(crimes),columns=crimes.columns)
```

```python
crimes1
```

```python
from pandas.core.common import random_state
x=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(crimes)
  x.append(kmeans.inertia_)

plt.plot(range(1,11),x)
plt.title('elbow')
plt.xlabel('clusters')
plt.ylabel('within cluster sum of squares')
plt.show()
```

```python
cluster_algorithm=KMeans(n_clusters=4,random_state=99)
cluster_algorithm.fit(crimes1)
```

```python
crimes1['clusters']=cluster_algorithm.labels_
```

```python
crimes1
```

```python
crimes1.groupby('clusters').agg(['mean']).reset_index()
```
