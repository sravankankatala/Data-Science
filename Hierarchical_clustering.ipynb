```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Airline=pd.read_excel('/EastWestAirlines.xlsx',sheet_name='data')
```

```python
Airline.head()
```

```python
Airline.info()
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
airline2=Airline.drop(['ID#'],axis=1)
```

```python
airline2.info()
```

```python
airline3=pd.DataFrame(normalize(airline2),columns=airline2.columns)
```

```python
airline3
```

```python
plt.figure(figsize=(20,7))
dendogram=sch.dendrogram(sch.linkage(airline3,method='complete'))
```

```python
Airlinecluster=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
Airlinecluster
```

```python
y=pd.DataFrame(Airlinecluster.fit_predict(airline3),columns=['clusterid'])
```

```python
y
```

```python
airline2['clusterid']=y
```
